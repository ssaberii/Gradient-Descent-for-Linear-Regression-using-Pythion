```python
# Gradient Descent for Linear Regression
```

```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Generate sample data
## Example: y = 2*x + 3 + noise
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 + 2 * X + np.random.randn(100, 1)
```

```python
# Step 3: Initialize parameters
# Random initial values for slope (theta1) and intercept (theta0)
theta0 = 0.0  # intercept
theta1 = 0.0  # slope
learning_rate = 0.1
iterations = 1000
m = len(y)  # number of data points
```

```python
# Step 4: Gradient Descent Algorithm
for i in range(iterations):
    # Predictions
    y_pred = theta0 + theta1 * X

    # Compute gradients
    d_theta0 = (1/m) * np.sum(y_pred - y)
    d_theta1 = (1/m) * np.sum((y_pred - y) * X)

    # Update parameters
    theta0 -= learning_rate * d_theta0
    theta1 -= learning_rate * d_theta1

    # Optional: print cost every 100 iterations
    if i % 100 == 0:
        cost = (1/(2*m)) * np.sum((y_pred - y)**2)
        print(f"Iteration {i}: Cost={cost}, theta0={theta0}, theta1={theta1}")
```

```python
# Step 5: Plot the result
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, theta0 + theta1 * X, color='red', label='Linear fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
